```python
months=['jan','feb','mar','apr','may','jun','jul','aug','sep','oct','nov','dec']

plt.plot(df["month_number"],df["total_profit"],color="green",marker='o',linestyle='--')
plt.xlabel("months")
plt.ylabel("units sold")
plt.title("total units sold over each months")
plt.xticks(range(1,13),months)
plt.show()
#total profit made in each month
```

```python
plt.plot(df['facecream'],marker='o',mfc='green')
plt.plot(df['facewash'],marker='s',mfc='red')
plt.plot(df['toothpaste'],marker='*',mfc='blue')
plt.plot(df['bathingsoap'],marker='^',mfc='yellow')
plt.plot(df['shampoo'],marker='x',mfc='orange')
plt.plot(df['moisturizer'],marker='.',mfc='pink')
plt.title("different items sold in each month")

plt.show()

#need another method to insert legend into the plot
```
