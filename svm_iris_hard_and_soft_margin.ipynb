```python
import numpy as np
import pandas as pd
import sklearn
import matplotlib.pyplot as plt
```

```python
from sklearn import datasets
```

```python
iris=datasets.load_iris()
```

```python
iris.data
```

```python
print(iris.target_names)
```

```python
x=iris.data[:,2:]
print(x)
```

```python
print(iris.feature_names)
```

```python
y=iris.target
y
```

```python
from sklearn.model_selection  import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler=StandardScaler()
x_train_scaled=scaler.fit_transform(x_train)
x_test_scaled=scaler.transform(x_test)
```

```python
x_train_scaled
```

```python
x_test_scaled
```

```python
from sklearn.svm import SVC
```

```python
#HARD MARGIN
svm_hard_margin=SVC(kernel="linear",C=0.1e9)
svm_hard_margin.fit(x_train_scaled,y_train)
```

```python
y_pred_hard_margin=svm_hard_margin.predict(x_test_scaled)
print(y_pred_hard_margin)
```

```python
from sklearn.metrics import accuracy_score
accuracy_hard_margin=accuracy_score(y_test,y_pred_hard_margin)
```

```python
print("Accuracy (hard margin svm):",accuracy_hard_margin)
```

```python
#SOFT MARGIN
svm_soft_margin=SVC(kernel="linear",C=0.1)
svm_soft_margin.fit(x_train_scaled,y_train)
```

```python
y_pred_soft_margin=svm_hard_margin.predict(x_test_scaled)
print(y_pred_soft_margin)
```

```python
print(y_test)
```

```python
accuracy_soft_margin=accuracy_score(y_test,y_pred_soft_margin)
print("Accuracy (SOFT margin svm):",accuracy_soft_margin)
```
